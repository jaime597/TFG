```python
import rmskmeans as km
import pandas as pd
import random
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import time

datafile = 'csv/dataset_final.csv'
fraction = 0.003
num_iter = 10
min_size = 100
num_cluster = 100
instance = 'def/clust'
```

```python
tic = time.process_time()
tstic = time.time()

#dataread = pd.read_csv(datafile, sep='|')[['paciente','utm30x_coord','utm30y_coord']].dropna()
#data = dataread.rename(columns={'paciente':'id', 'utm30x_coord':'x','utm30y_coord':'y'})



data = pd.read_csv(datafile, sep='|').rename(columns={'paciente':'id', 'utm30x_coord':'x','utm30y_coord':'y'})
km.rmskmeans(data, num_cluster, min_size=min_size, n_iter=num_iter, fraction=fraction, images=True, name=instance)

toc = time.process_time()
tstoc = time.time()
with open('%s.met'%(instance), 'w') as f:
    print('execution name: ', instance, end='\n', file=f)
    print('algorithm: [parallel] randomized min size kmeans (PRMS kmeans)', instance, end='\n', file=f)
    print('num. clusters: ', num_cluster, end='\n', file=f)
    print('min cluster size: ', min_size, end='\n', file=f)
    print('dataset size : ', len(data), end='\n', file=f)
    print('sample fraction : ', fraction, end='\n', file=f)
    print('# iterations: ', num_iter, end='\n', file=f)
    print('process_time: %f'%(toc - tic), end='\n', file=f)     
    print('execution_time: %f'%(tstoc - tstic))     

print('process_time: %f'%(toc - tic))     
print('execution_time: %f'%(tstoc - tstic))     

f.close()
```
